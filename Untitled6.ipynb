```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
#
```

```python
df = pd.read_csv('satislar.csv')
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['Aylar'], df['Satislar'], test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train.reshape(-1,1)) # Remove .values and reshape
x_test = sc.fit_transform(x_test.reshape(-1,1)) # Remove .values and reshape
y_train = sc.fit_transform(y_train.values.reshape(-1,1)) # Reshape y_train to 2D
y_test = sc.fit_transform(y_test.values.reshape(-1,1)) # Reshape y_test to 2D
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
prediction = lr.predict(x_test)
prediction
```

```python
sbn.regplot(x=x_test, y=y_test, ci=None, scatter_kws={'color':'r', 's':9})
```
